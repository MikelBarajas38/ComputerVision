```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Load Image

```python
img = cv2.imread('in/space_heavy.jpg')

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plt.imshow(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

# get grayscale image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('grayscale', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/gray.jpg', gray)

# get binary image
ret, thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
cv2.imshow('binary image', thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/binary.jpg', thresh)
```

### Bilateral Filter

```python
#bilateral filter
# params: source, diameter of each pixel neighborhood, sigmaColor, sigmaSpace

bf = cv2.bilateralFilter(img, 9, 75, 75)
cv2.imshow('original', img)
cv2.imshow('bilateral filter', bf)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/bilateral.jpg', bf)

bin_bf = cv2.bilateralFilter(thresh, 9, 75, 75)
cv2.imshow('binary image', thresh)
cv2.imshow('bilateral filter', bin_bf)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/bin_bilateral.jpg', bin_bf)
```

### Blur

```python
# blur
# params: source, kernel size, anchor

blur = cv2.blur(img, (5,5))
cv2.imshow('original', img)
cv2.imshow('blur', blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/blur.jpg', blur)

bin_blur= cv2.blur(thresh, (5,5))
cv2.imshow('original', thresh)
cv2.imshow('blur', bin_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/bin_blur.jpg', bin_blur)
```

### Dilate

```python
# dilate
# params: source, kernel, iterations

dilate = cv2.dilate(img, (5,5), iterations = 3)
cv2.imshow('original', img)
cv2.imshow('dilate', dilate)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/dilate.jpg', dilate)

dilate_bin = cv2.dilate(thresh, (5,5), iterations = 3)
cv2.imshow('original', thresh)
cv2.imshow('dilate', dilate_bin)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/dilate_bin.jpg', dilate_bin)
```

### Erode

```python
# erode
# params: source, kernel, iterations

erode = cv2.erode(img, (5,5), iterations = 3)
cv2.imshow('original', img)
cv2.imshow('erode', erode)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/erode.jpg', erode)

erode_bin = cv2.erode(thresh, (5,5), iterations = 3)
cv2.imshow('original', thresh)
cv2.imshow('erode', erode_bin)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/erode_bin.jpg', erode_bin)
```

### Gaussian Blur

```python
# gaussian blur
# params: source, kernel size, sigmaX

gblur = cv2.GaussianBlur(img, (5,5), 0)
cv2.imshow('original', img)
cv2.imshow('gaussian blur', gblur)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/gblur.jpg', gblur)

gblur_bin = cv2.GaussianBlur(thresh, (5,5), 0)
cv2.imshow('original', thresh)
cv2.imshow('gaussian blur', gblur_bin)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/gblur_bin.jpg', gblur_bin)
```

### Median Blur

```python
# median blur
# params: source, kernel size

mblur = cv2.medianBlur(img, 5)
cv2.imshow('original', img)
cv2.imshow('median blur', mblur)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/mblur.jpg', mblur)

mblur_bin = cv2.medianBlur(thresh, 5)
cv2.imshow('original', thresh)
cv2.imshow('median blur', mblur_bin)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/mblur_bin.jpg', mblur_bin)
```

### Spatial Gradient

```python
# spatial gradient
# params: source

grad_x, grad_y = cv2.spatialGradient(gray)
cv2.imshow('original', gray)
cv2.imshow('gradient x', grad_x)
cv2.imshow('gradient y', grad_y)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('out/grad_x.jpg', grad_x)
cv2.imwrite('out/grad_y.jpg', grad_y)

magnitude = np.sqrt((grad_x ** 2) + (grad_y ** 2))
orientation = np.arctan2(grad_y, grad_x) 

cv2.imshow('gradient magnitude', magnitude)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/grad_mag.jpg', magnitude)

cv2.imshow('gradient orientation', orientation)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('out/grad_or.jpg', orientation)


# equalize histogram

equ = cv2.equalizeHist(gray)

# spatial gradient
# params: source

grad_x, grad_y = cv2.spatialGradient(equ)
cv2.imshow('original', equ)
cv2.imshow('gradient x', grad_x)
cv2.imshow('gradient y', grad_y)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('out/equ.jpg', equ)
cv2.imwrite('out/equ_grad_x.jpg', grad_x)
cv2.imwrite('out/equ_grad_y.jpg', grad_y)

```
