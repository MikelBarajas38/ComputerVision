```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Load Image

```python
img = cv2.imread('hands.jpg')

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plt.imshow(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Change image to grayscale

```python
img = cv2.imread('hands.jpg')

cv2.imshow('original image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# parameters: source, color space conversion code
grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('grayscale image', grayscale_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('grayscale_hands.jpg', grayscale_img)
```

## Histogram Equalization

```python
# get the histogram of the grayscale image

grayscale_hist = cv2.calcHist([grayscale_img], [0], None, [256], [0, 256])

plt.title('grayscale histogram')
plt.xlabel('intensity')
plt.ylabel('# of pixels')
plt.xlim([0, 256])
plt.plot(grayscale_hist, color='k')
# plt.hist(grayscale_img.ravel(), 256, [0, 256])
plt.show()

# get the cumulative distribution function of the grayscale image

cdf = grayscale_hist.cumsum()
plt.title('grayscale cumulative distribution function')
plt.xlabel('intensity')
plt.ylabel('# of pixels')
plt.xlim([0, 256])
plt.plot(cdf, color='k')
plt.show()

# equalize the histogram

equalized_grayscale_img = cv2.equalizeHist(grayscale_img)

cv2.imshow('equalized grayscale image', equalized_grayscale_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('equalized_grayscale_hands.jpg', equalized_grayscale_img)

# get the histogram of the equalized grayscale image

equalized_grayscale_hist = cv2.calcHist([equalized_grayscale_img], [0], None, [256], [0, 256])

plt.title('equalized grayscale histogram')
plt.xlabel('intensity')
plt.ylabel('# of pixels')
plt.xlim([0, 256])
plt.plot(equalized_grayscale_hist, color='k')
# plt.hist(equalized_grayscale_img.ravel(), 256, [0, 256])
plt.show()

# get the cumulative distribution function of the equalized grayscale image

equalized_cdf = equalized_grayscale_hist.cumsum()
plt.title('equalized grayscale cumulative distribution function')
plt.xlabel('intensity')
plt.ylabel('# of pixels')
plt.xlim([0, 256])
plt.plot(equalized_cdf, color='k')
plt.show()
```

```python
# change image to YrCrCb color space

ycrcb_img = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

cv2.imshow('YrCrCb image', ycrcb_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('ycrcb_hands.jpg', ycrcb_img)

# get the histograms of the YrCrCb image

y_hist = cv2.calcHist([ycrcb_img], [0], None, [256], [0, 256])

plt.title('YrCrCb histogram')
plt.xlabel('intensity')
plt.ylabel('# of pixels')
plt.xlim([0, 256])
plt.plot(y_hist, color='k')
plt.show()

# get the cumulative distribution function of the YrCrCb image

y_cdf = y_hist.cumsum()

plt.title('YrCrCb cumulative distribution function')
plt.xlabel('intensity')
plt.ylabel('# of pixels')
plt.xlim([0, 256])
plt.plot(y_cdf, color='k')
plt.show()

# equalize the YrCrCb image

equalized_ycrcb_img = ycrcb_img.copy()
equalized_ycrcb_img[:, :, 0] = cv2.equalizeHist(ycrcb_img[:, :, 0])

cv2.imshow('equalized YrCrCb image', equalized_ycrcb_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('equalized_ycrcb_hands.jpg', equalized_ycrcb_img)

# get the histograms of the equalized YrCrCb image

equalized_y_hist = cv2.calcHist([equalized_ycrcb_img], [0], None, [256], [0, 256])

plt.title('equalized YrCrCb histogram')
plt.xlabel('intensity')
plt.ylabel('# of pixels')
plt.xlim([0, 256])
plt.plot(equalized_y_hist, color='k')
plt.show()

# get the cumulative distribution function of the equalized YrCrCb image

equalized_y_cdf = equalized_y_hist.cumsum()

plt.title('equalized YrCrCb cumulative distribution function')
plt.xlabel('intensity')
plt.ylabel('# of pixels')
plt.xlim([0, 256])
plt.plot(equalized_y_cdf, color='k')
plt.show()

```

```python
# get channel images from YrCrCb image

y, cr, cb = cv2.split(ycrcb_img)

cv2.imshow('Y channel', y)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('Cr channel', cr)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('Cb channel', cb)
cv2.waitKey(0)
cv2.destroyAllWindows()

# get cool and warm color images from YrCrCb image

cool_img = ycrcb_img.copy()
cool_img[:, :, 0] = 128
cool_img[:, :, 2] = 128

cv2.imshow('cool image', cv2.cvtColor(cool_img, cv2.COLOR_YCrCb2BGR))
cv2.waitKey(0)
cv2.destroyAllWindows()

warm_img = ycrcb_img.copy()
warm_img[:, :, 0] = 128
warm_img[:, :, 1] = 128

cv2.imshow('warm image', cv2.cvtColor(warm_img, cv2.COLOR_YCrCb2BGR))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
