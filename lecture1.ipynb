## Contrast & Brightness

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('img/test.jpg')

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plt.imshow(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

# convertTo() function
img_contrast_high_2 = cv2.convertScaleAbs(img, alpha = 2, beta = 0)

cv2.imshow('contrast a = 2',img_contrast_high_2)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plt.imshow(img)
plt.title('contrast a = 2')
plt.imshow(cv2.cvtColor(img_contrast_high_2, cv2.COLOR_BGR2RGB))
plt.show()

# convertTo() function
img_contrast_high_4 = cv2.convertScaleAbs(img, alpha = 0.25, beta = 0)

cv2.imshow('contrast a = 0.25',img_contrast_high_4)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plt.imshow(img)
plt.title('contrast a = 0.25')
plt.imshow(cv2.cvtColor(img_contrast_high_4, cv2.COLOR_BGR2RGB))
plt.show()
```

## Histogram Equalization

```python
#https://docs.opencv.org/3.4/d8/dbc/tutorial_histogram_calculation.html

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('img/test.jpg')

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plt.imshow(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

hist_eq_img = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

cv2.imshow('YCrCb',hist_eq_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plt.imshow(img)
plt.imshow(cv2.cvtColor(hist_eq_img, cv2.COLOR_BGR2RGB))
plt.show()

# split the image into channels
channels = cv2.split(hist_eq_img)

# applying histogram equalisation to the Y channel
cv2.equalizeHist(channels[0], channels[0])

# merge the channels
hist_eq_img = cv2.merge(channels)

# converting back to BGR format
hist_eq_img = cv2.cvtColor(hist_eq_img, cv2.COLOR_YCrCb2BGR)

cv2.imshow('Histogram Equalisation',hist_eq_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plt.imshow(img)
plt.imshow(cv2.cvtColor(hist_eq_img, cv2.COLOR_BGR2RGB))
plt.show()

```
